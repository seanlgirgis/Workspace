### 🛠️ The Mathematical Model

Let $f(c, S)$ be a function that returns the frequency of character $c$ in string $S$.
$S$ and $T$ are anagrams if for every character $c$ in $S \cup T$:

$$f(c, S) = f(c, T)$$

---

### 📋 Problem

Given two strings `s` and `t`, return `true` if `t` is an anagram of `s`, and `false` otherwise.

**Example 1:**
```
Input:  s = "anagram", t = "nagaram"
Output: true
```

**Example 2:**
```
Input:  s = "rat", t = "car"
Output: false
```

**Constraints:** `s` and `t` consist of lowercase English letters.

## 🐢 Approach 1: Sort First — $O(N \log N)$

````python
def isAnagram_sort(s, t):
    """
    Sort Elements
    Time: O(N log N) | Space: O(N) or O(1) depending on language
    """
    return sorted(s) == sorted(t)


# Test Cases
print(isAnagram_sort("anagram", "nagaram"))  # Expected: True
print(isAnagram_sort("rat", "car"))          # Expected: False
````

## 🚀 The Optimal Approach: $O(N)$

By maintaining a manual frequency map (`count = {}`), we directly measure the elements in exactly one pass over each array.

### The Key Lifecycle Rule
1. **Check if lengths match.** If they don't, return early. $O(1)$ constraint.
2. **Add occurrences** from `s` to the hash map.
3. **Subtract occurrences** found in `t` from the hash map.
4. **Validate** that no character counter drops below zero.

---

## ✅ Mathematical Proof

$$Time \approx O(N) + O(N) = O(2N) \rightarrow O(N)$$

<div style="padding: 10px; border-left: 8px solid #4CAF50; background-color: #f1f8f4; color: #1b5e20; border-radius: 4px;">
    <strong>✅ Summary:</strong> Evaluating linear loops explicitly bypasses sequence manipulation overhead completely resolving parameter checks dynamically.
</div>

## 🚀 Approach 2: Hash Map Counter (Optimal) — $O(N)$
---

Instead of sorting objects sequentially mapping array indexes, we utilize **Dictionaries** as an **accounting ledger**.

As we iterate:
1. Map `s` elements explicitly adding frequencies.
2. Map `t` operations explicitly decrementing frequencies dynamically verifying sequence boundaries.
3. Determine if limits inherently violate equal equivalence bounding constraints securely.

````python
def isAnagram_optimized(s, t):
    """
    Optimal: Manual Frequency Hash Map
    Time: O(N) | Space: O(1) (Bounded by 26 English characters)
    """
    if len(s) != len(t):
        return False
        
    count = {}
    
    for char in s:
        count[char] = count.get(char, 0) + 1
        
    for char in t:
        if char not in count or count[char] == 0:
            return False
        count[char] -= 1
        
    return True


print("Optimal:", isAnagram_optimized("anagram", "nagaram"))  # Expected: True
print("Optimal:", isAnagram_optimized("rat", "car"))          # Expected: False
````
