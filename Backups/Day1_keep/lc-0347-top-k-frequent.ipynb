### 🛠️ The Mathematical Model

Let $C(v)$ be the frequency of an element $v$ in array $A$ of length $N$.
The maximum possible frequency for any element is bounding exactly to $N$.
By defining an array `buckets` of size $N+1$, we guarantee $C(v)$ correctly maps as a valid index:

$$0 \leq C(v) \leq N$$

### 📋 Problem

Output: [1,2]
```

**Example 2:**
```

## 🐢 Approach 1: Sort by Dictionary Keys — $O(N \log N)$

````python
from collections import Counter

def topKFrequent_sort(nums, k):
    """
    Sort Elements
    Time: O(N log N) | Space: O(N)
    """
    count = Counter(nums)
    # Sort the dictionary keys based on their values descending
    return sorted(count.keys(), key=count.get, reverse=True)[:k]


# Test Cases
print(topKFrequent_sort([1, 1, 1, 2, 2, 3], 2))  # Expected: [1, 2]
print(topKFrequent_sort([1], 1))                 # Expected: [1]
````

````python
def topKFrequent_optimized(nums, k):
    """
    Optimal: Bucket Sort Frequency Indexing
    Time: O(N) | Space: O(N)
    """
    count = Counter(nums)
    
    # Frequency cannot exceed len(nums). Initialize N + 1 buckets.
    buckets = [[] for _ in range(len(nums) + 1)]
    
    for num, freq in count.items():
        buckets[freq].append(num)
        
    res = []
    for i in range(len(buckets) - 1, 0, -1):
        for num in buckets[i]:
            res.append(num)
            if len(res) == k:
                return res
                
    return []


print("Optimal:", topKFrequent_optimized([1, 1, 1, 2, 2, 3], 2))  # Expected: [1, 2]
print("Optimal:", topKFrequent_optimized([1], 1))                 # Expected: [1]
````
