### 🛠️ The Mathematical Model

Given that we need to find `nums[i] + nums[j] == target`.
By rearranging the formula, `nums[j] == target - nums[i]`.
We can calculate the required complement instantly instead of guessing.

$$nums_j = Target - nums_i$$

---

### 📋 Problem

Given an array of integers `nums` and an integer `target`, return indices of the two numbers such that they add up to `target`.

**Example 1:**
```
Input:  nums = [2, 7, 11, 15], target = 9
Output: [0, 1]
```

**Example 2:**
```
Input:  nums = [3, 2, 4], target = 6
Output: [1, 2]
```

**Constraints:** Exactly one solution exists. An element cannot be used twice.

## 🐢 Approach 1: Brute Force — $O(N^2)$

````python
def twoSum_brute(nums, target):
    """
    Brute Force: Check every combination via nested iterations.
    Time: O(N^2) | Space: O(1)
    """
    n = len(nums)
    for i in range(n):
        for j in range(i + 1, n):
            if nums[i] + nums[j] == target:
                return [i, j]
    return []


# Test Cases
print(twoSum_brute([2, 7, 11, 15], 9))  # Expected: [0, 1]
print(twoSum_brute([3, 2, 4], 6))  # Expected: [1, 2]
````

## 🔬 Complexity Analysis: $O(N^2)$ vs. $O(N)$

<div style="padding: 15px; border-left: 8px solid #2196F3; background-color: #f0f7ff; color: #0d47a1; border-radius: 4px;">
    <strong>The Core Insight:</strong> A naive nested iteration examines elements it has already checked in the past.
</div>

---

## 📉 Why Brute Force Fails: The $O(N^2)$ Trap

A nested iteration runs exactly $N * (N-1) / 2$ times. As the array grows to 100,000 items, the CPU is running nearly 5 Billion operations.

$$\frac{N(N-1)}{2}$$

| Input Type | Brute Force Performance | Reason |
|------------|------------------------|--------|
| **Worst-Case Ending** | N * (N - 1) / 2 | The matching values are the final two items in the array. |
| **Large Arrays (10,000+)** | Millions of CPU cycles | A sequential scan prevents rapid index lookup. |

---

## 🚀 The Optimal Approach: $O(N)$

By maintaining a dictionary (`seen = {}`), we drastically optimize lookup processing to $O(1)$. It's a complete paradigm shift: "calculate and look" instead of "guess and check."

### The Key Lifecycle Rule
1. **Check if calculation was seen**
2. **Add to dictionary if seen was False**

---

## ✅ Mathematical Proof

$$Time \approx O(N) \times O(1)$$

<div style="padding: 10px; border-left: 8px solid #4CAF50; background-color: #f1f8f4; color: #1b5e20; border-radius: 4px;">
    <strong>✅ Summary:</strong> N elements are evaluated exactly once, resulting in a strictly linear execution path.
</div>

## 🚀 Approach 2: One-Pass HashMap (Optimal) — $O(N)$
---

Instead of checking every following number, we use **a Hash Dictionary** as a **backward-looking cache**.

As we iterate:
1. Compute the exact `complement` value needed.
2. Check if it exists in the dictionary cache.
3. Add the value to the dictionary cache if not found.

````python
def twoSum_optimized(nums, target):
    """
    Optimal: One-Pass HashMap
    Time: O(N) | Space: O(N)
    """
    seen = {}
    
    for i, num in enumerate(nums):
        complement = target - num
        
        if complement in seen:
            return [seen[complement], i]
            
        seen[num] = i
        
    return []


print("Optimal:", twoSum_optimized([2, 7, 11, 15], 9))  # Expected: [0, 1]
print("Optimal:", twoSum_optimized([3, 2, 4], 6))  # Expected: [1, 2]
````

## 🎤 5 Interview Q&A

### Q1: What is the space complexity and why?

**Answer:** Space is $O(N)$ because the dictionary `seen` could potentially store every single element up until the final index in the worst-case scenario.

---

### Q2: Is a dictionary lookup truly $O(1)$?

**Answer:** Dict queries perform an average of $O(1)$. In rare worst-case paths (hash collisions), python traverses a node chain scaling it linearly to $O(N)$. However, Python utilizes highly resilient hash tables utilizing pseudorandom numbers designed specifically for mitigation so the average resolves definitively as constant.

---

### Q3: What is the time complexity of the optimal approach and why?

**Answer:** $O(N)$. We loop through the elements linearly using a single pass for-loop mapping length $N$. Inside the loop, hash dictionary retrieval yields an independent execution timing of $O(1)$.

---

### Q4: If the array was already sorted, what technique could we transition into to reduce space constraint?

**Answer:** We could integrate a Two-Pointer technique initializing `left=0` and `right=len-1`. Space requirements convert exclusively out of dynamic arrays into single-variable states mapping us to $O(1)$ space with maintaining linear $O(N)$ runtime.

---

### Q5: What are the edge cases to watch for?

**Answer:** Duplicates within the input sequence mapping to the target output. Executing map assignment unconditionally at the loop origin allows subsequent index elements to overwrite keys containing duplicate scalar content causing execution path validation to fail when cross-mapping matches elements against their identical twins.
