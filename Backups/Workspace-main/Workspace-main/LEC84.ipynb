            <div class="concept-card">
                <h2>📊 The Problem</h2>
                <p>Given an array of bar heights, find the largest rectangular area that can be formed.</p>

                <p><strong>Goal:</strong> Find the maximum area rectangle that fits within the histogram.</p>
            </div>

            <!-- Concept 2: Key Insight -->
            <div class="concept-card">
                <h2>💡 Key Insight</h2>
                <p>Instead of checking all possible rectangles (O(n²) or O(n³)), use a <strong>monotonic stack</strong> to keep track of potential rectangles.</p>
                
                <div class="key-insight">
                    <strong>The Stack Metaphor:</strong><br>
                    Think of bars as customers in a "waiting room." Each bar waits for a shorter bar to arrive. When a shorter bar comes, the waiting customers (taller bars) can't extend further—we calculate their final area.
                </div>

                <p><strong>Why this works:</strong></p>
                <ul style="margin-left: 20px; color: #555; line-height: 1.8;">
                    <li>Stack maintains increasing heights</li>
                    <li>When we hit a shorter bar, we know how far back the taller bars can extend</li>
                    <li>Time complexity: O(n) instead of O(n²)</li>
                </ul>
            </div>
        </div>

        <!-- Algorithm Steps -->
        <div class="algorithm-section">
            <h2>🔧 How the Algorithm Works</h2>
            
            <div class="step">
                <span class="step-number">1</span>
                <strong>Maintain a Stack</strong>
                <p>Store pairs of (index, height) for bars with increasing heights. This forms a "waiting line" of bars that haven't encountered a blocker yet.</p>
            </div>

            <div class="step">
                <span class="step-number">2</span>
                <strong>Process Each Bar</strong>
                <p>For each new bar, if it's shorter than the bar on top of the stack, we've found a "blocker." The stacked bars can't extend further right.</p>
            </div>

            <div class="step">
                <span class="step-number">3</span>
                <strong>Calculate Areas</strong>
                <p>When popping a bar from the stack, we know:</p>
                <ul style="margin-left: 40px; margin-top: 10px;">
                    <li><strong>Height:</strong> The popped bar's height</li>
                    <li><strong>Right boundary:</strong> The current bar's index</li>
                    <li><strong>Left boundary:</strong> Where the popped bar originally started (stored in stack)</li>
                </ul>
            </div>

            <div class="step">
                <span class="step-number">4</span>
                <strong>Cleanup</strong>
                <p>After processing all bars, any remaining bars in the stack were never blocked. They can extend all the way to the right edge. Calculate their areas with the right boundary = array length.</p>
            </div>
        </div>

        <!-- Detailed Mechanics -->
        <div class="concept-grid">
            <div class="concept-card">
                <h2>🎯 The Stack Trick</h2>
                <p><strong>When we pop a bar:</strong> We don't just know its height—we know exactly how far it can extend!</p>
                
                <div class="stack-container">
                    Stack: [(idx=0, h=2), (idx=1, h=4)]<br>
                    Current bar: height=3 at index=2
                </div>

                <p style="margin-top: 15px;"><strong>Calculation for popped bar (idx=1, h=4):</strong></p>
                <div class="stack-container">
                    • Right boundary: current index = 2<br>
                    • Left boundary: top of remaining stack = 0<br>
                    • Width: 2 - 0 - 1 = 1<br>
                    • Area: 4 × 1 = 4
                </div>

                <p style="margin-top: 15px; color: #666; font-size: 0.9em;"><em>The "start" variable tracks where a bar could have started extending left from previous pops.</em></p>
            </div>

            <div class="concept-card">
                <h2>⚡ Why It's Optimal</h2>
                
                <div class="comparison-grid">
                    <div class="comparison-box suboptimal">
                        <div class="comparison-title">❌ Brute Force</div>
                        <div class="comparison-content">
                            Check every possible rectangle combination<br><br>
                            Time: O(n²) or O(n³)<br>
                            Space: O(1)
                        </div>
                    </div>
                    <div class="comparison-box optimal">
                        <div class="comparison-title">✅ Stack Approach</div>
                        <div class="comparison-content">
                            Each bar processed once, pushed/popped once<br><br>
                            Time: O(n)<br>
                            Space: O(n)
                        </div>
                    </div>
                </div>

                <p style="margin-top: 15px; color: #555; line-height: 1.6;">The stack approach is <strong>optimal</strong> because it processes each bar exactly twice (push + pop), making it linear time.</p>
            </div>
        </div>

        <!-- Code Visualization -->
        <div class="algorithm-section">
            <h2>📝 Annotated Code Logic</h2>
            
            <div class="code-block"><span class="keyword">def</span> largestRectangleArea(heights):
    max_area = <span class="string">0</span>
    stack = []  <span class="comment"># (starting_index, height) pairs</span>
    
    <span class="keyword">for</span> index, height <span class="keyword">in</span> enumerate(heights):
        start = index
        <span class="comment"># Bar is shorter: finalize rectangles</span>
        <span class="keyword">while</span> stack <span class="keyword">and</span> stack[-<span class="string">1</span>][<span class="string">1</span>] > height:
            popped_index, popped_height = stack.pop()
            <span class="comment"># Width = from where it started to current blocker</span>
            width = index - popped_index
            max_area = max(max_area, popped_height * width)
            start = popped_index  <span class="comment"># Extend left for next bar</span>
        
        <span class="comment"># Push current bar (or extended version)</span>
        stack.append((start, height))
    
    <span class="comment"># Cleanup: bars that were never blocked</span>
    <span class="keyword">for</span> index, height <span class="keyword">in</span> stack:
        max_area = max(max_area, height * (len(heights) - index))
    
    <span class="keyword">return</span> max_area</div>

            <div class="key-insight">
                <strong>Magic Line:</strong> <code>start = popped_index</code><br>
                This lets shorter bars "inherit" the starting position of taller bars. A bar of height 3 can extend as far left as the original position of a previously-popped height-5 bar.
            </div>
        </div>

        <!-- Example Trace -->
        <div class="algorithm-section">
            <h2>🔍 Example Walkthrough</h2>
            <p style="color: #555; margin-bottom: 20px;"><strong>Input:</strong> heights = [2, 4, 3, 2, 4]</p>
            
            <div style="display: grid; gap: 15px;">
                <div class="step">
                    <strong>Index 0, height=2:</strong><br>
                    Stack is empty → Push (0, 2)<br>
                    Stack: [(0, 2)]
                </div>

                <div class="step">
                    <strong>Index 1, height=4:</strong><br>
                    4 > 2 → No popping → Push (1, 4)<br>
                    Stack: [(0, 2), (1, 4)]
                </div>

                <div class="step">
                    <strong>Index 2, height=3:</strong><br>
                    3 < 4 → Pop (1, 4): Area = 4 × (2 - 1) = 4<br>
                    3 > 2 → Stop popping<br>
                    Push (1, 3) <span style="color: #999;">[bar-2 inherits bar-1's start]</span><br>
                    Stack: [(0, 2), (1, 3)]
                </div>

                <div class="step">
                    <strong>Index 3, height=2:</strong><br>
                    2 < 3 → Pop (1, 3): Area = 3 × (3 - 1) = 6<br>
                    2 = 2 → Stop popping<br>
                    Push (1, 2) <span style="color: #999;">[bar-3 inherits start=1]</span><br>
                    Stack: [(0, 2), (1, 2)]
                </div>

                <div class="step">
                    <strong>Index 4, height=4:</strong><br>
                    4 > 2 → No popping → Push (4, 4)<br>
                    Stack: [(0, 2), (1, 2), (4, 4)]
                </div>

                <div class="step">
                    <strong>Cleanup (end of array):</strong><br>
                    Pop (4, 4): Area = 4 × (5 - 4) = 4<br>
                    Pop (1, 2): Area = 2 × (5 - 1) = 8 ← <strong>Maximum!</strong><br>
                    Pop (0, 2): Area = 2 × (5 - 0) = 10 ← <strong>Actually maximum!</strong>
                </div>
            </div>

            <p style="margin-top: 20px; color: #555;"><strong>Answer: 10</strong> (height-2 bar extends across entire width)</p>
        </div>
