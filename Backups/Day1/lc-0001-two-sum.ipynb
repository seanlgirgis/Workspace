# 🧠 LeetCode 1: Two Sum

**Pattern:** Hashing / Array

- **Created:** 2026-02-28
- **Focus:** Fast dictionary lookups, matching pairs
- **Tags:** `array` | `hash-table` | `easy` | `citi-prep`

## 📖 Problem Statement

Given an array of integers `nums` and an integer `target`, return indices of the two numbers such that they add up to `target`.

You may assume that each input would have exactly one solution, and you may not use the same element twice. You can return the answer in any order.

### Example 1:
- **Input:** `nums = [2,7,11,15]`, `target = 9`
- **Output:** `[0,1]`
- **Explanation:** Because `nums[0] + nums[1] == 9`, we return `[0, 1]`.

## 🧠 Mental Model & Intuition

Instead of looking at a number and asking "who can I pair with?", look at a number and ask *"who do I urgently need?"* (The Diff).

If the target is 9, and you are 2, you *need* a 7. 
We keep a ledger (Hash Map) of everyone we've seen so far. 
If the number we *need* is in the ledger, we found our pair!

## 🐢 Brute Force Approach

Check every possible pair with nested loops.

```python
def twoSumBruteForce(nums, target):
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] + nums[j] == target:
                return [i, j]
```
**Time:** $O(n^2)$ | **Space:** $O(1)$

````python
def twoSum(nums: list[int], target: int) -> list[int]:
    # Value -> Index mapping
    seen = {}
    
    for i, num in enumerate(nums):
        diff = target - num
        
        if diff in seen:
            return [seen[diff], i]
            
        # Otherwise, add ourselves to the ledger for future numbers
        seen[num] = i
        
    return []

# Test
print(twoSum([2, 7, 11, 15], 9)) # Expected: [0, 1]
````

## ⏱️ Complexity Analysis

- **Time Complexity:** $O(N)$ because we traverse the array exactly once. Hash map internal lookups are average $O(1)$.
- **Space Complexity:** $O(N)$ because in the worst case (the pair is at the very end), we store $N-1$ elements in the hash map.

## 🎤 Interview Q&A

### Q1: What if the array is sorted?
**Answer:** I would skip the Hash Map entirely and use a Two-Pointer approach (one at index 0, one at index N-1). That would drop the Space Complexity to $O(1)$ while keeping Time at $O(N)$.

---

### Q2: Why check the hash map *before* adding the current element?
**Answer:** To prevent an element from matching with itself. For example, if target is 6 and the array is `[3, 2, 4]`, adding `3` first might make the loop think it found a match at the same index `0`.

## 📚 Key Terminology

| Term | Definition | Example |
|---|---|---|
| **Hash Map** | A data structure mapping keys to values for $O(1)$ lookups. | Python `dict` |
| **Complement (Diff)** | The value needed to satisfy the condition alongside the current value. | `target - current` |
| **Ledger Pattern** | Keeping track of "seen" items to enable $O(1)$ retrospective checks in a loop. | `seen[value] = index` |

## 💼 The Citi Narrative

**Context:** Trading reconcilements.

**Scenario:** At Citi, we had a ledger of buy trades and a ledger of sell trades. To reconcile breaks (unmatched trades), we had to find one buy and one sell that netted out to a specific dollar discrepancy. Matching them with nested loops brought the database to a crawl during end-of-day batches.

**Impact:** Loading the buys into memory as a Python Dictionary (amount -> trade_id) allowed us to linear-scan the sells and lookup the exact complementing difference in $O(1)$ time, reducing the script execution from 45 minutes to 3 seconds.

````python
# EXERCISE: Modify Two Sum to return ALL unique pairs that sum to target

def twoSumAllPairs(nums, target):
    # To do this, we need a set to track elements instead of just storing one index.
    seen = set()
    pairs = set()
    
    for num in nums:
        diff = target - num
        if diff in seen:
            # Sort pair so (2,7) is same as (7,2)
            pairs.add(tuple(sorted([num, diff])))
        seen.add(num)
    return list(pairs)

print(twoSumAllPairs([2, 7, 11, 15, 7, 2], 9))
````

## 🎯 Summary: Key Takeaways

### The Pattern
**Hash Map (One-Pass)** — Storing complements in a hash map for O(1) lookups during iteration.

### When to Use It
- ✅ Finding pairs that match a target condition.
- ✅ O(1) temporal lookups of previously seen streaming states.
- ❌ **Don't use when:** Need continuous ranges or sub-arrays (Use Sliding Window instead).
