# 🧠 LeetCode 242: Valid Anagram

## 📖 Problem Statement

Given two strings `s` and `t`, return `true` if `t` is an anagram of `s`, and `false` otherwise.

An **Anagram** is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

### Example 1:
- **Input:** `s = "anagram"`, `t = "nagaram"`
- **Output:** `true`

### Example 2:
- **Input:** `s = "rat"`, `t = "car"`
- **Output:** `false`

## 🧠 Mental Model & Intuition

Imagine two buckets of Lego bricks. To know if they can build the exact same house, you don't need to try building the house.

You just count the bricks. If bucket A has 5 red, 3 blue, and 2 yellow bricks, and bucket B has exactly the same counts, then B is an anagram of A.

We use a Hash Map (or fixed-size array) as our "tally sheet".

## 🐢 Brute Force Approach

Sort both strings and compare them.

```python
def isAnagramBrute(s: str, t: str) -> bool:
    return sorted(s) == sorted(t)
```
**Time:** $O(N \log N)$ (Due to sorting) | **Space:** $O(N)$ (Or $O(1)$ if the language allows in-place string sorting, but Python strings are immutable)

````python
# Optimal Approach: Hash Map Frequency Counter
def isAnagram(s: str, t: str) -> bool:
    if len(s) != len(t):
        return False
        
    countS, countT = {}, {}
    
    # Build the tally sheets
    for i in range(len(s)):
        countS[s[i]] = countS.get(s[i], 0) + 1
        countT[t[i]] = countT.get(t[i], 0) + 1
        
    # Now comparing the dicts directly in Python does a deep equality check
    # return countS == countT
    
    # To be explicit for an interview:
    for char in countS:
        if countS[char] != countT.get(char, 0):
            return False
            
    return True

print("Is 'anagram' an anagram of 'nagaram'?", isAnagram("anagram", "nagaram"))
````

````python
# An EVEN MORE optimal approach if we know inputs are lowercase English letters (a-z).
# A fixed array of size 26 avoids Hash Map overhead.
def isAnagramArray(s: str, t: str) -> bool:
    if len(s) != len(t): return False
    
    counts = [0] * 26
    for i in range(len(s)):
        # The ord() function gets the ASCII value. ord('a') is 97.
        # 'c' (99) - 'a' (97) -> index 2.
        counts[ord(s[i]) - ord('a')] += 1
        counts[ord(t[i]) - ord('a')] -= 1
        
    for val in counts:
        if val != 0:
            return False
            
    return True

print("Is 'rat' an anagram of 'car'?", isAnagramArray("rat", "car"))
````

## ⏱️ Complexity Analysis

- **Time Complexity:** $O(N)$ where $N$ is the length of the strings. We iterate over the strings a constant number of times.
- **Space Complexity:** $O(1)$ if assuming the character set is fixed (e.g., 26 lowercase English letters, or even 256 Extended ASCII). The hash map or array never grows beyond this fixed constant.

## 🎤 Interview Q&A

### Q1: What if the inputs contain Unicode characters (like emojis)?
**Answer:** The fixed-size array approach of size 26 won't work. We would revert to using a standard Hash Map (dictionary) because it can dynamically track any Unicode point as a key.

---

### Q2: Why check `len(s) != len(t)` first?
**Answer:** An incredibly cheap $O(1)$ check that prunes impossible branches immediately. If the lengths differ, there's zero chance they are anagrams. Never skip this check.

## 💼 The Citi Narrative

**Context:** Schema Validation.

**Scenario:** In the Equities division, upstream systems would send configuration files in XML. Sometimes, developers would accidentally scramble the order of the configuration tags. I wrote a validator script.

**Impact:** To quickly verify that `File B` contained the exactly same set of tags as `File A` (just in a different order), I used the Frequency Tally pattern. I tallied all tag occurrences in A, subtracted the occurrences in B, and ensured the array values were all zero. It was instantaneous, avoiding heavy XML parsing comparisons.

````python
# EXERCISE: Use Python's built-in Counter library (which does exactly this under the hood)
from collections import Counter

def isAnagramPythonic(s: str, t: str) -> bool:
    return Counter(s) == Counter(t)

print(isAnagramPythonic("listen", "silent"))
````

## 🎯 Summary: Key Takeaways

### The Pattern
**Frequency Counter** — Using an array or map to exact-match character counts between two sequences.

### When to Use It
- ✅ Checking permutations or identical letter compositions.
- ✅ Fast distribution equality checks.
- ❌ **Don't use when:** Sequences where order matters (Use direct string comparison).
