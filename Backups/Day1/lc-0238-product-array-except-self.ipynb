## 📖 Problem Statement

Given an integer array `nums`, return an array `answer` such that `answer[i]` is equal to the product of all the elements of `nums` except `nums[i]`.

The product of any prefix or suffix of `nums` is guaranteed to fit in a 32-bit integer.

**Constraint:** You must write an algorithm that runs in $O(N)$ time and **without using the division operation**.

### Example 1:
- **Input:** `nums = [1,2,3,4]`
- **Output:** `[24,12,8,6]`

## 🧠 Mental Model & Intuition

If we *could* use division, we would multiply everything together ($1\times2\times3\times4 = 24$), then map the array dividing by the item: ($24/1, 24/2, 24/3, 24/4$). But we can't because of limits on zeros and problem rules.

Imagine you are at position `i`. The product of everything *except you* is exactly equal to the product of everything to your **left**, multiplied by the product of everything to your **right**.

So, we build two lists:
1. `prefix`: the cumulative product going -> left to right.
2. `postfix`: the cumulative product going <- right to left.

Then for any slot `i`, the answer is `prefix[i-1] * postfix[i+1]`.

## 🐢 Brute Force Approach

For every number, loop through the whole array again to multiply everything but the current index.

```python
def productExceptSelfBrute(nums):
    res = []
    for i in range(len(nums)):
        prod = 1
        for j in range(len(nums)):
            if i != j:
                prod *= nums[j]
        res.append(prod)
    return res
```
**Time:** $O(N^2)$ (Causes Timeout) | **Space:** $O(N)$ for result

````python
# Optimal Approach: Prefix and Postfix arrays (Space optimized to O(1) extra space)
def productExceptSelf(nums: list[int]) -> list[int]:
    res = [1] * len(nums)
    
    # 1. First Pass: Compute everything to the LEFT (Prefix)
    prefix = 1
    for i in range(len(nums)):
        res[i] = prefix
        prefix *= nums[i]
        
    # Now res holds all the left products.
    # res[i] = mult(nums[0..i-1])
    
    # 2. Second Pass: Compute everything to the RIGHT (Postfix)
    postfix = 1
    for i in range(len(nums) - 1, -1, -1):
        # we multiply the stored Left product by the growing Right product
        res[i] *= postfix
        postfix *= nums[i]
        
    return res

print("Output for [1,2,3,4]:", productExceptSelf([1, 2, 3, 4]))
````

## ⏱️ Complexity Analysis

- **Time Complexity:** $O(N)$ because we make exactly two linear sweeps across the array. $O(2N) = O(N)$.
- **Space Complexity:** $O(1)$ auxiliary memory (the output array `res` does not count as extra memory for space complexity analysis in LeetCode).

## 🎤 Interview Q&A

### Q1: Why not just use division as a fallback in reality?
**Answer:** In the real world, I absolutely would—unless the array contains zeros. If an array has one zero, dividing by it throws a `ZeroDivisionError`. If it has two zeros, the product of the whole array is 0, rendering the division trick useless. The prefix/suffix pattern is immune to zero issues.

---

### Q2: How did you achieve O(1) space?
**Answer:** Conceptually, this algorithm requires an array for the Left products and an array for the Right products. By storing the Left products directly in the output `res` array, and tracking the Right product dynamically with a single `postfix` integer during the reverse sweep, we eliminate the need for extra structure allocations.

## 📚 Key Terminology

| Term | Definition | Example |
|---|---|---|
| **Prefix Array** | An array where index $i$ stores an aggregation (sum/product) of all elements before $i$. | `prefix[i]` |
| **Postfix Array** | The reverse; an aggregation from the end to $i+1$. | `postfix[i]` |
| **Two-Pass Algorithm** | Solving a problem by looping through the data entirely forwards, then backwards. | Doing `prefix`, then `postfix` sweep. |

## 💼 The Citi Narrative

**Context:** Market impact simulation.

**Scenario:** Risk analysts needed to calculate a "shock modifier" for a basket of 10 related bonds. The modifier for Bond A meant multiplying the volatility indices of Bonds B through J (everything except A). The division approach failed spectacularly every time one bond's volatility hit 0.0, causing the entire modeling pipeline to crash.

**Impact:** Implementing a Prefix/Postfix multiplication sweep in the Python pricing engine fixed the zero-division bugs permanently and made the algorithm mathematically robust against outlier data points in volatile markets.

## 🎯 Summary: Key Takeaways

### The Pattern
**Prefix / Suffix Arrays** — Pre-computing cumulative products from both directions to avoid division.

### When to Use It
- ✅ Range queries where state depends on everything 'before' and 'after'.
- ✅ Avoiding division by zero operations in aggregate metrics.
- ❌ **Don't use when:** Queries that only look sequentially in one direction.
