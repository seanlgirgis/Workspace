<strong>The Problem:</strong> You are planning to rob houses along a street. Each house has a certain amount of money stashed. The only constraint blocking you from robbing each of them is that adjacent houses carry connected security systems. Given an integer array `nums` representing the amount of money of each house, return the maximum amount you can rob without triggering alarms.
</div>

**Example:**
```text
Input: nums = [1,2,3,1]
Output: 4
Explanation: Rob house 1 (money = 1) and then rob house 3 (money = 3). Total = 4.
```


### Core Insight
At any given house `i`, you have a localized choice: either rob it (adding its value, but forfeiting house `i-1`), OR skip it entirely (keeping whatever max profits you accrued all the way through house `i-1`). Structurally: `max_profit(i) = max(rob_current + max_profit(i-2), skip_current_assume_i-1)`.

## Mental Models

**1. The Recurrence Relation**
```python
dp[i] = max(nums[i] + dp[i-2], dp[i-1])
```
If you rob `i`, you get `nums[i]` plus the max profit possible by the time you finished evaluating `i-2`. If you skip `i`, the total profit naturally mirrors whatever your max was arriving at `i-1`.

**2. The Moving Window**
Identical explicitly to Climbing Stairs! Calculating house `i` relies purely on `i-1` and `i-2`. An array is entirely gratuitous. Using two shifting temporal pointers securely collapses memory profiling mathematically perfectly.

````python
# Test Suite
def run_tests(fn):
    assert fn([1, 2, 3, 1]) == 4, "Base Example 1"
    assert fn([2, 7, 9, 3, 1]) == 12, "Base Example 2 (2+9+1)"
    assert fn([2, 1, 1, 2]) == 4, "Edge: Needs contiguous gap (2, _, _, 2)"
    assert fn([5]) == 5, "Edge: One house"
    assert fn([5, 1, 1, 5]) == 10, "Edge: Gapping two spaces entirely"
    print("✅ All tests passed!")
````

````python
# Brute Force (Recursive)
# O(2^n) time | O(n) space

def rob_brute(nums: list[int]) -> int:
    def dfs(i: int) -> int:
        if i < 0: return 0
        return max(nums[i] + dfs(i - 2), dfs(i - 1))
    return dfs(len(nums) - 1)
````

````python
# Optimal (Dynamic Programming - Constant Space)
# O(n) Time | O(1) Space

def rob(nums: list[int]) -> int:
    if not nums: return 0
    
    rob1, rob2 = 0, 0
    
    # [rob1, rob2, n, n+1, ...] moving the window forward
    for n in nums:
        temp = max(n + rob1, rob2)
        rob1 = rob2
        rob2 = temp
        
    return rob2

run_tests(rob)
````

## Interview Q&A
