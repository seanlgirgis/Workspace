<strong>The Problem:</strong> You are climbing a staircase. It takes `n` steps to reach the top. Each time you can either climb 1 or 2 steps. In how many distinct ways can you climb to the top?
</div>

**Example:**
```text
Input: n = 3
Output: 3
Explanation: There are three ways to climb to the top.
1. 1 step + 1 step + 1 step
2. 1 step + 2 steps
3. 2 steps + 1 step
```


### Core Insight
To reach step `n`, you either took a 1-step from `n-1`, or a 2-step from `n-2`. Thus, the total ways to reach step `n` is exactly the sum of ways to reach `n-1` and ways to reach `n-2`. `ways(n) = ways(n-1) + ways(n-2)`. This is the Fibonacci sequence in disguise!

## Mental Models

**1. Base Cases & Induction**
Step 1: 1 way (1). Step 2: 2 ways (1+1, 2). Step 3: `ways(2) + ways(1) = 3`.

**2. Tree of Decisions**
Every step spans out two branches (+1 and +2). Drawing the tree exposes immense overlapping subproblems (e.g. recalculating step 3 dozens of times for higher `n`). Recursion solves the tree, Memoization/DP flattens it into O(N).

**3. Constant Space Tradeoff**
Since we only ever need the last two values computationally, an array of size N is wasteful. Simply use two integer variables shifting positions.

````python
# Test Suite
def run_tests(fn):
    assert fn(2) == 2, "Base case 2"
    assert fn(3) == 3, "Base case 3"
    assert fn(4) == 5, "Recursive step 4 (1+1+1+1, 1+1+2, 1+2+1, 2+1+1, 2+2)"
    assert fn(5) == 8, "Recursive step 5"
    print("✅ All tests passed!")
````

````python
# Brute Force (Recursive)
# O(2^n) time | O(n) space

def climbStairs_brute(n: int) -> int:
    if n == 1: return 1
    if n == 2: return 2
    return climbStairs_brute(n - 1) + climbStairs_brute(n - 2)

# run_tests(climbStairs_brute)  # Fails at n=45 due to extreme slowness
````

````python
# Optimal (Dynamic Programming - Constant Space)
# O(n) Time | O(1) Space

def climbStairs(n: int) -> int:
    if n <= 2:
        return n
        
    prev2 = 1  # ways to reach n-2
    prev1 = 2  # ways to reach n-1
    
    for i in range(3, n + 1):
        current = prev1 + prev2
        prev2 = prev1
        prev1 = current
        
    return prev1

run_tests(climbStairs)
````

## Complexity Analysis

| Approach | Time | Space | Notes |
|----------|------|-------|-------|
| Recursive | O(2ⁿ) | O(n) | Complete binary tree traversal. |
| DP (Array) | O(n) | O(n) | Explicit memoization structure. |
| **DP (Optimized)** | **O(n)** | **O(1)** | Standard acceptable optimal. |
| Matrix Exp | O(log n) | O(1) | Esoteric. Not expected in interview unless math-centric. |

## Interview Q&A

**Q: What is overlapping subproblems?**
A: In the naive tree, to solve `F(5)`, `F(3)` is computed heavily on the left branch, and then completely re-computed entirely again on the right branch. DP guarantees a subproblem is evaluated exactly once by saving the state.

**Q: Can we solve this top-down?**
A: Yes, Memoization (passing an array and conditionally returning array[n] if physically calculated) yields Top-Down O(N) Time and Space. The loop mechanism yields Bottom-Up DP. Bottom-Up safely secures O(1) space heavily preferred in algorithmic interviews.

**Q: Does this work for 1, 2, or 3 increments?**
A: Trivially. `current = prev1 + prev2 + prev3`. Standard scaling dynamic.

**Q: Python specific: Would `@cache` from `functools` work flawlessly?**
A: Yes, placing `@cache` securely atop the naive recursive physically memoizes function signatures converting O(2^n) immediately to O(n) functionally. Great trick if you freeze under pressure.
