<strong>The Problem:</strong> You are given an integer array `coins` representing coins of different denominations and an integer `amount` representing a total amount of money. Return the fewest number of coins that you need to make up that amount. If that amount of money cannot be made up by any combination of the coins, return `-1`.
You can use any coin any number of times.
</div>

**Example:**
```text
Input: coins = [1,2,5], amount = 11
Output: 3
Explanation: 11 = 5 + 5 + 1
```


### Core Insight
This is the quintessential "knapsack-style" DP. Think of building it bottom-up. To find the minimum coins for amount `11`, if I have a `[5]` coin, I look at the minimum coins previously calculated for `amount - 5 = 6`, and just add 1 (my new 5-coin). `dp[a] = min(dp[a], 1 + dp[a - c])` over all valid coins `c`.

## Mental Models

**1. The Infinity Array**
Initialize an array of size `amount + 1` with a "dummy infinity" value like `amount + 1`. You can never use more than `amount` coins (worst case 1-cent coins), so `amount + 1` safely acts as infinity. `dp[0]` is explicitly 0.

**2. Look-Back Strategy**
For every single amount from 1 up to `amount`, test every single coin. "If I stand at total X, and look back exactly `coin` steps, what was the minimum score there? + 1". We take the absolute minimum of all possible coin look-backs.

**3. Why Greedy Fails**
Always taking the biggest coin first *fails* on arbitrary denominations. E.g. `coins=[1, 3, 4]`, `amount=6`. Greedy takes `4`, then `1, 1` (3 coins). DP optimally takes `3, 3` (2 coins).

````python
# Test Suite
def run_tests(fn):
    assert fn([1, 2, 5], 11) == 3, "Base Example (5, 5, 1)"
    assert fn([2], 3) == -1, "Impossible amount inherently"
    assert fn([1], 0) == 0, "Zero amount functionally returns 0 explicitly"
    assert fn([1, 3, 4], 6) == 2, "Greedy breaker (requires 3, 3 instead of 4, 1, 1)"
    assert fn([186, 419, 83, 408], 6249) == 20, "Large iteration boundary"
    print("✅ All tests passed!")
````

````python
# Brute Force (Recursive Top-Down — Unbounded Branching)
# O(C^A) time | O(A) space strictly (Max depth A levels)

def coinChange_brute(coins: list[int], amount: int) -> int:
    def dfs(rem: int) -> int:
        if rem == 0: return 0
        if rem < 0: return float('inf')
        
        min_cost = float('inf')
        for coin in coins:
            res = dfs(rem - coin)
            if res != float('inf'):
                min_cost = min(min_cost, 1 + res)
        return min_cost
        
    res = dfs(amount)
    return res if res != float('inf') else -1

# Fails natively via TLE instantly on high depths
````

````python
# Optimal (Dynamic Programming Bottom-Up)
# O(A * C) Time | O(A) Space
# A = amount linearly, C = len(coins)

def coinChange(coins: list[int], amount: int) -> int:
    # Initialize dp array to 'infinity' (amount + 1 is natively mathematically impossible to exceed)
    dp = [amount + 1] * (amount + 1)
    dp[0] = 0
    
    for a in range(1, amount + 1):
        for c in coins:
            if a - c >= 0:
                dp[a] = min(dp[a], 1 + dp[a - c])
                
    return dp[amount] if dp[amount] != amount + 1 else -1

run_tests(coinChange)
````

## Complexity Analysis

| Approach | Time | Space | Notes |
|----------|------|-------|-------|
| Pure Greedy | O(C log C) | O(1) | Completely Incorrect Results! |
| Memoized DFS | O(A * C) | O(A) | Hits python max recursion depths natively. |
| **Bottom-Up DP** | **O(A * C)** | **O(A)** | Safely bounds loop mapping explicit O(A) memory grids. |

## Interview Q&A

**Q: Why do we explicitly initiate arrays strictly to `amount + 1` natively?**
A: Because mathematically finding the `min()` necessitates the placeholder start infinitely high. Float Infinity creates casting issues occasionally in native languages. `amount + 1` literally acts identically because the maximum conceivable physical coins for `amount` is exclusively `amount` (solely pennies!).

**Q: What exactly isolates Coin Change conceptually from regular permutations?**
A: The inner loop natively tests `c` directly upon the linear `a` loop independently resulting in explicitly calculating "Combinations", not ordered permutations. It evaluates the exact minimal threshold mathematically without path dependence.

**Q: In what specific constrained scenarios purely does Greedy inherently succeed?**
A: The canonical US Coin System! (1, 5, 10, 25). If standard coin denominations uniquely guarantee every coin is strictly mathematically greater than the sum of all subsequent maximum available multiples natively dynamically Greedy actually works identically O(1) space.
