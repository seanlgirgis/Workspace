<strong>The Problem:</strong> Given an integer array `nums`, return the length of the longest strictly increasing subsequence.
</div>

**Example:**
```text
Input: nums = [10,9,2,5,3,7,101,18]
Output: 4
Explanation: The longest increasing subsequence is [2,3,7,101], therefore the length is 4.
```


### Core Insight
A subsequence doesn't have to be contiguous! To find the max LIS ending at index `i`, we natively examine *every preceding* index `j` `< i`. If `nums[i] > nums[j]`, then we fundamentally can securely append our `nums[i]` explicitly mathematically extending `j`'s sequence length by `+1`.

## Mental Models

**1. The "Ending Minimum" Default Array**
Initialize an array `dp` of identically matched length to `nums` entirely initialized strictly to `1`. Every explicit single number inherently structurally operates as a valid subsequence identically natively mapping to length exactly `1` unconditionally.

**2. Double Loop Backward Look**
For every `i`, scan every `j` from `0` to `i-1`. Compare: if I natively strictly append myself to `j`, what's my sequence length? (`dp[j] + 1`). Is it strictly mathematically greater evaluating `dp[i]` natively? If yes, overwrite `dp[i]`.

**3. Global Maximums**
The longest sequence naturally does NOT inherently mandate terminating flawlessly exclusively ending identically at the ultimate extremely final integer. Iterating the absolute `max(dp)` globally intrinsically guarantees detecting the peak.

````python
# Test Suite
def run_tests(fn):
    assert fn([10, 9, 2, 5, 3, 7, 101, 18]) == 4, "Standard Mixed Values (2,3,7,101 or 2,5,7,18)"
    assert fn([0, 1, 0, 3, 2, 3]) == 4, "Multiple overlapping drops"
    assert fn([7, 7, 7, 7, 7, 7, 7]) == 1, "Strictly identical requires explicitly increasing ONLY"
    print("✅ All tests passed!")
````

````python
# Brute Force (Recursive Top-Down)
# O(2^n) time | O(n) space

def lengthOfLIS_brute(nums: list[int]) -> int:
    def dfs(i: int, prev_val: float) -> int:
        if i == len(nums): return 0
        
        # Option 1: Skip entirely purely naturally
        res = dfs(i + 1, prev_val)
        
        # Option 2: Strictly append identically confirming mathematically explicitly ascending
        if nums[i] > prev_val:
            res = max(res, 1 + dfs(i + 1, nums[i]))
            
        return res

    return dfs(0, float('-inf'))
````

````python
# Optimal (Dynamic Programming Bottom-Up Array)
# O(n^2) Time | O(n) Space

def lengthOfLIS(nums: list[int]) -> int:
    if not nums: return 0
    
    # Base case: every element is a sequence of len 1 natively strictly
    dp = [1] * len(nums)
    
    for i in range(len(nums)):
        for j in range(i):
            if nums[i] > nums[j]:
                dp[i] = max(dp[i], dp[j] + 1)
                
    return max(dp)

run_tests(lengthOfLIS)
````
