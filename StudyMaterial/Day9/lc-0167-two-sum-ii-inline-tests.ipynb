# LC #167 — Two Sum II (Input Array is Sorted) [Medium]
**Category:** Two Pointers

### Problem
Given a **sorted** array `numbers` (1-indexed), find two numbers that add to `target`. Return `[index1, index2]` (1-indexed). Use only O(1) extra space.
`numbers = [2, 7, 11, 15]`, `target = 9` → `[1, 2]`

### 🎯 Core Insight
The array is sorted. If the sum is too big, which pointer moves? If too small?
Two pointers is the pattern for problems on sorted arrays where you need pairs, triples, or a range. The trick: sorting creates structure. From a sorted array, you can move pointers intelligently — the left pointer moves right when the sum is too small, right pointer moves left when too large. This drops O(n²) brute force to O(n) after the O(n log n) sort. The sliding window IS a two-pointer pattern — the difference is that two pointers usually work from both ends while sliding window works from one end.

### 🧠 Mental Models
1. **The Squeeze:** "I start at the extremes. If the pairing is too heavy, I shrink the max. If too light, I raise the min."
2. **Hash Map Fallback:** Although not optimal for space, you could record what you've seen. If the complement exists in the record, you found the pair. We avoid this to maintain O(1) space.

```python
def run_tests(test_cases, func):
    print(f"\n{'='*40}\nTesting: {func.__name__}\n{'='*40}")
    passed = 0
    for i, (args, expected) in enumerate(test_cases):
        result = func(*args)
        if result == expected:
            print(f"  ✅ Test {i+1} PASSED")
            passed += 1
        else:
            print(f"  ❌ Test {i+1} FAILED: Expected {expected}, Got {result}")
    print(f"\n{passed}/{len(test_cases)} tests passed.\n{'='*40}\n")
```

```python
# Brute Force (Conceptual)
# Time: O(n^2), Space: O(1)

def twoSum_brute(numbers: list[int], target: int) -> list[int]:
    for i in range(len(numbers)):
        for j in range(i + 1, len(numbers)):
            if numbers[i] + numbers[j] == target:
                return [i + 1, j + 1]
    return []

test_cases = [
    (([2, 7, 11, 15], 9), [1, 2]),
    (([2, 3, 4], 6), [1, 3]),
    (([-1, 0], -1), [1, 2])
]

run_tests(test_cases, twoSum_brute)
```

```python
# Optimal: Two Pointers
# Time: O(n), Space: O(1)

def twoSum(numbers: list[int], target: int) -> list[int]:
    left, right = 0, len(numbers) - 1

    while left < right:
        total = numbers[left] + numbers[right]
        if total == target:
            return [left + 1, right + 1]    # 1-indexed
        elif total < target:
            left += 1    # need bigger sum → move left right
        else:
            right -= 1   # need smaller sum → move right left

    return []

test_cases = [
    (([2, 7, 11, 15], 9), [1, 2]),
    (([2, 3, 4], 6), [1, 3]),
    (([-1, 0], -1), [1, 2])
]

run_tests(test_cases, twoSum)
```

### 📊 Complexity Analysis
- **Time Complexity:** O(n). Each pointer moves at most n steps total. No nested loops — every iteration eliminates at least one element from consideration.
- **Space Complexity:** O(1). We only use two pointers, independent of input size.
*(If we used a Hash Map, it would be O(n) time and O(n) space. Brute force would be O(n²) time).*

### 🎤 Interview Q&A
- **Q: Why does this work?** 
  A: Array is sorted. If sum < target, only option is to increase it → move left right (larger values). If sum > target, decrease it → move right left. We never skip a valid pair.
- **Q: Why is this O(n) instead of O(n²)?** 
  A: Each pointer moves at most n steps total. No nested loops — every iteration eliminates at least one element from consideration.
- **Q: What happens if the array is NOT sorted?** 
  A: This approach fails because moving pointers doesn't guarantee structured sum changes. You'd need to sort it first (O(n log n)) or use a Hash Map (O(n) time, O(n) space).
- **Q: Are there edge cases?** 
  A: The problem guarantees exactly one solution. If there were duplicates, moving pointers naturally skims through them, or we could explicitly skip duplicates if finding unique pairs.
