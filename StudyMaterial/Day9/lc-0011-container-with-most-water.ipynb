# LC #11 — Container With Most Water [Medium]
**Category:** Two Pointers (Greedy)

### Problem
Given heights `height[]`, find two lines that together with the x-axis form a container holding the most water.
`height = [1,8,6,2,5,4,8,3,7]` → `49`

### 🎯 Core Insight
Width decreases as pointers move in. The only way to increase area is to find a taller line. So always move the shorter pointer inward.

### 🧠 Mental Models
1. **The Greedy Funnel:** "I start with the widest possible bucket. If I shrink the width, I MUST find a taller wall to compensate. I drop the shortest wall because it limits my capacity."
2. **Max Envelope:** "The area is bounded by the smallest limit. Squeezing the boundaries only makes sense if the limit increases."

```python
def run_tests(test_cases, func):
    print(f"\n{'='*40}\nTesting: {func.__name__}\n{'='*40}")
    passed = 0
    for i, (args, expected) in enumerate(test_cases):
        result = func(*args)
        if result == expected:
            print(f"  ✅ Test {i+1} PASSED")
            passed += 1
        else:
            print(f"  ❌ Test {i+1} FAILED: Expected {expected}, Got {result}")
    print(f"\n{passed}/{len(test_cases)} tests passed.\n{'='*40}\n")
```

```python
# Brute Force (Conceptual)
# Time: O(n^2), Space: O(1)

def maxArea_brute(height: list[int]) -> int:
    max_water = 0
    n = len(height)
    for i in range(n):
        for j in range(i + 1, n):
            width = j - i
            current = min(height[i], height[j]) * width
            max_water = max(max_water, current)
    return max_water

test_cases = [
    (([1,8,6,2,5,4,8,3,7],), 49),
    (([1,1],), 1),
    (([4,3,2,1,4],), 16),
    (([1,2,1],), 2)
]

run_tests(test_cases, maxArea_brute)
```

```python
# Optimal: Two Pointers (Greedy)
# Time: O(n), Space: O(1)

def maxArea(height: list[int]) -> int:
    left, right = 0, len(height) - 1
    max_water = 0

    while left < right:
        width = right - left
        current = min(height[left], height[right]) * width
        max_water = max(max_water, current)

        # Move the shorter side — the taller side can't help with a smaller width
        if height[left] <= height[right]:
            left += 1
        else:
            right -= 1

    return max_water

test_cases = [
    (([1,8,6,2,5,4,8,3,7],), 49),
    (([1,1],), 1),
    (([4,3,2,1,4],), 16),
    (([1,2,1],), 2)
]

run_tests(test_cases, maxArea)
```

### 📊 Complexity Analysis
- **Time Complexity:** O(n). We use two pointers that start at opposite ends and move inward until they meet. The code inside the `while` loop runs exactly once per element.
- **Space Complexity:** O(1). Only a few integer variables (`left`, `right`, `max_water`, `width`, `current`) are created to store state.

### 🎤 Interview Q&A
- **Q: Why move the shorter pointer?**
  A: Area = min(left_height, right_height) × width. When we move inward, width decreases by 1. The only hope of increasing area is finding a taller line. Moving the taller line guarantees we keep or decrease the height — no upside. Moving the shorter line gives a chance to find something taller.
- **Q: Is this provably optimal?**
  A: Yes — by induction. If we move the shorter pointer, we've already computed the best area possible for that shorter line at any position (since any position to the right has ≤ width). No valid answer is skipped.
- **Q: What happens if both heights are exactly equal?**
  A: You can safely move either pointer (or even both). Since the capacity is bottlenecked by both equal heights, finding a taller line on just one side won't increase the area because the opposite equal-height line will remain the bottleneck while the width shrinks.
- **Q: What if there is a massive spike in the middle of the array?**
  A: The two pointers will naturally converge over it. Since they constantly discard the smaller outer walls, they will climb inward until they both stand on the massive spikes (if there are two), extracting the maximum valid area.
- **Q: Can this be solved using Divide and Conquer?**
  A: Technically yes, but it would involve evaluating cross-boundary containers which is complex and usually bounds out at O(n log n). The greedy two-pointer strategy at O(n) is the strictly superior approach for this problem.
