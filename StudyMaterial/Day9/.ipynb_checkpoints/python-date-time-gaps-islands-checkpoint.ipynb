## Quick Reference Card — The Time-Series Toolkit

```python
# ---------------------------------------------------------
# 1. Pandas Vectorized Gaps & Islands (The Fast Way)
# ---------------------------------------------------------
# Flag where the date difference is NOT 1 day
df['is_new_island'] = df['date'].diff().dt.days != 1
# → [True, False, False, True, False]

# Cumulative sum of the boolean mask creates a unique Island ID!
df['island_id'] = df['is_new_island'].cumsum()
# → [1, 1, 1, 2, 2]

# Group by the new Island ID to find start, end, and duration
islands = df.groupby('island_id').agg({'date': ['min', 'max', 'count']})
# → returns consecutive streaks

# ---------------------------------------------------------
# 2. Finding Actual Gaps (The Missing Days)
# ---------------------------------------------------------
# Generate a perfect continuous "Date Spine"
spine = pd.date_range(start='2026-02-01', end='2026-02-28')
# → DatetimeIndex(['2026-02-01', '2026-02-02', ...])

# Find what is in the spine but missing from your data (Set Difference)
missing_dates = set(spine.date) - set(df['date'].dt.date)
# → {datetime.date(2026, 2, 5), datetime.date(2026, 2, 8)}
```

**Rule:** Use `itertools.groupby` for pure Python/streaming use cases (low memory) where data is already sorted. Use Pandas `cumsum()` vectorization for high-performance in-memory analytical datasets.

````python
# Code Demo 1: Performance Concept Proof (Memory Efficiency)
import sys
import itertools
import pandas as pd
import numpy as np

N = 1_000_000

# 1. Pandas Vectorized Allocation
# Allocates the full boolean mask and integer array in memory immediately.
# We use seconds frequency ('s') here because adding 1,000,000 days crashes Pandas' Year 2262 limit!
dates_series = pd.Series(pd.date_range("2020-01-01", periods=N, freq="s"))
island_ids_pd = (dates_series.diff().dt.total_seconds() != 1).cumsum()
print(f"Pandas Vectorized Grouping ({N:,} items): {sys.getsizeof(island_ids_pd):>11,} bytes  ← allocated instantly")

# 2. Itertools Generator (Lazy Grouping)
# Evaluates consecutive logical groupings strictly yielding one at a time
def lazy_date_generator(n):
    for i in range(n):
        yield "2020-01-01"

gen_group = itertools.groupby(lazy_date_generator(N))
print(f"Pure Python itertools Generator ({N:,}): {sys.getsizeof(gen_group):>11,} bytes  ← lazy constant memory")

print("\nAt 1M items: Pandas scales into Megabytes for raw speed.")
print("At 100M items: Pandas might OOM; itertools safely streams using only Bytes.")
````

## ⚙️ The Mechanics: How Vectorized Islands Work

In Pandas, we avoid row-by-row `for` loops. To detect Islands (consecutive blocks of days), we utilize boolean masking combined with a cumulative sum.

**Execution model:**
1. **Sort Phase:** Enforce that the DataFrame is explicitly sorted by `server_id` and `date`.
2. **Diff Phase:** `df['date'].diff().dt.days` calculates the integer day gap between the current row and the row directly above it.
3. **Mask Phase:** `!= 1` evaluates to `True` whenever a sequence breaks (a gap larger than 1 day, or a shift to a new server).
4. **Cumsum Phase:** `cumsum()` treats `True` as `1` and `False` as `0`. A rolling sum of these flags produces a stair-step integer array where every identical integer mathematically represents a unique continuous Island!

````python
# Code Demo 2: Finding Islands in Mock Server Data
import io
import pandas as pd

SAMPLE_CSV = """server_id,cpu_pct,date,tier
srv-01,72,2026-02-01,prod
srv-01,73,2026-02-02,prod
srv-01,95,2026-02-03,prod
srv-01,97,2026-02-06,prod
srv-01,99,2026-02-07,prod
srv-02,45,2026-02-01,dev
"""

df = pd.read_csv(io.StringIO(SAMPLE_CSV))
df['date'] = pd.to_datetime(df['date'])
print("—— Raw Server Metrics ——")
print(df)

df = df.sort_values(['server_id', 'date']).reset_index(drop=True)

# 1. Calculate days since prev row. Shift servers logic so diff doesn't bridge servers.
df['day_diff'] = df.groupby('server_id')['date'].diff().dt.days

# 2. Flag rows ending an island (diff != 1 means sequence broken)
df['is_new_island'] = df['day_diff'] != 1

# 3. Cumsum creates the unique Island Keys
df['island_id'] = df['is_new_island'].cumsum()

# 4. Group by to summarize the streaks
islands = df.groupby(['server_id', 'island_id']).agg(
    streak_start=('date', 'min'),
    streak_end=('date', 'max'),
    streak_days=('date', 'count')
).reset_index()

print("\n—— Identified Islands (Consecutive Blocks) ——")
print(islands[['server_id', 'island_id', 'streak_start', 'streak_end', 'streak_days']])
````

````python
# Code Demo 3: Pipeline Showcase — Finding Gaps using Lazy Generators
# Pipeline: parse → filter_dates → enrich_spine → materialize_gaps

import io
from collections import defaultdict

def parse_csv_stream(fileobj):
    next(fileobj)
    for line in fileobj:
        p = line.strip().split(',')
        yield {'server_id': p[0], 'date': p[2]}

def map_to_servers(record_stream):
    observed = defaultdict(set)
    for rec in record_stream:
        observed[rec['server_id']].add(rec['date'])
    return observed  # Yields materialized state

def calculate_gaps(server_map, date_spine_set):
    for srv, dates_seen in server_map.items():
        # Set difference instantly finds the missing holes
        missing = date_spine_set - dates_seen
        yield srv, missing

print("—— Lazy Gap Pipeline Results ——")
f = io.StringIO(SAMPLE_CSV)
expected_spine = {f"2026-02-{day:02d}" for day in range(1, 8)}

# The Chain
parsed_stream = parse_csv_stream(f)
server_dict = map_to_servers(parsed_stream)
gaps_stream = calculate_gaps(server_dict, expected_spine)

for srv, missing_set in gaps_stream:
    print(f"{srv} missed polling on dates: {missing_set}")

print("\nPipeline executed using constant memory lazy generators.")
````

## Functional Date Tooling — Standard Library & Pandas Hooks

When integrating sequence parsing, `itertools.groupby` combined with Date Spines (`pd.date_range`) gives you unparalleled programmatic leverage:

````python
import pandas as pd
from itertools import groupby

# 1. pd.date_range — Generating the master date spine natively
spine = pd.date_range('2026-03-01', periods=3)
print("1. pd.date_range:\n", spine, "\n")
# → DatetimeIndex(['2026-03-01', '2026-03-02', '2026-03-03']...)

# 2. itertools.groupby — Grouping by a dynamic boolean threshold
# This groups consecutive blocks natively without pandas overhead.
data_points = [1, 2, 3, 10, 11, 4, 5]  # Values under 10 are valid
print("2. itertools.groupby (Finding Islands purely in Python):")
for is_valid, group in groupby(data_points, key=lambda x: x < 10):
    print(f"  Island {is_valid}: {list(group)}")
# → Island True: [1, 2, 3]; Island False: [10, 11]; Island True: [4, 5]

# 3. df.reindex — Conforming fragmented data to a Date Spine
s = pd.Series(["A", "B"], index=pd.to_datetime(['2026-03-01', '2026-03-03']))
filled = s.reindex(spine)
print("\n3. df.reindex (Creating Gaps gracefully):\n", pd.DataFrame(filled, columns=['val']))
# → NaN introduced on 2026-03-02
````

## Interview Q&A

**Q: Explain the Pandas logic for locating Islands.**
A: You first sort the dataframe by Entity and Date. Then, invoke `.diff().dt.days` to find gaps larger than 1 day from the preceding row. Finally, taking the `.cumsum()` of this boolean mask creates a unique integer identifier for every unbroken consecutive streak (the Island).
**Q: What is a Date Spine and why is it required for finding Gaps?**
A: A Date Spine is a perfectly contiguous generated array of dates (using `pd.date_range()`). It's required because you cannot directly query something that physically does not exist in logs; by comparing against a perfect Date Spine via Set Differences or Outer Joins, the missing Gaps explicitly materialize as `NaN` or unmatched records.
**Q: Why use `itertools.groupby` instead of Pandas `.groupby()` for streaming data?**
A: Pandas requires allocating the entire dataset into memory to conduct a grouped aggregation. `itertools.groupby` operates lazily, processing one sequential chunk at a time. It's the only option when file sizes drastically exceed system RAM.
**Q: What is the cardinal rule of using `itertools.groupby`?**
A: You MUST sort the data by the target key before feeding it to `itertools.groupby`. Unlike SQL, the Python standard library `groupby` only groups contiguous adjoining items; if the stream is randomized, you will get severely fragmented groupings.
**Q: How do you identify the longest continuous streak within Pandas?**
A: After creating your `island_id` via `cumsum()`, you run a `df.groupby(['server_id', 'island_id']).size()`. Sorting those size outputs descending instantly yields the longest consecutive duration.
