# LC #42 — Trapping Rain Water [Hard]
**Category:** Two Pointers (Prefix Max)

### Problem
Given heights `height[]`, compute total water trapped after rain.
`height = [0,1,0,2,1,0,1,3,2,1,2,1]` → `6`

### 🎯 Core Insight
Water at each position = min(max height to the left, max height to the right) - height at position. Two pointers track these maximums in O(1) space, moving inward from the boundaries.

### 🧠 Mental Models
1. **The Safe Bet:** "If I know there's a huge wall somewhere on my right, my water level is entirely bottlenecked by the tallest wall I've seen on my left. I don't need to know exactly how tall the right wall is, just that it's taller than my left cap."
2. **Prefix/Suffix Bounds (DP approach):** "Create a map of the absolute highest peak looking left, and the highest looking right. The water above me is the minimum of those two peaks minus my ground elevation."

```python
def run_tests(test_cases, func):
    print(f"\n{'='*40}\nTesting: {func.__name__}\n{'='*40}")
    passed = 0
    for i, (args, expected) in enumerate(test_cases):
        result = func(*args)
        if result == expected:
            print(f"  ✅ Test {i+1} PASSED")
            passed += 1
        else:
            print(f"  ❌ Test {i+1} FAILED: Expected {expected}, Got {result}")
    print(f"\n{passed}/{len(test_cases)} tests passed.\n{'='*40}\n")
```

```python
# Suboptimal DP Approach
# Time: O(n), Space: O(n)

def trap_dp(height: list[int]) -> int:
    if not height: return 0

    n = len(height)
    left_max = [0] * n
    right_max = [0] * n
    
    left_max[0] = height[0]
    for i in range(1, n):
        left_max[i] = max(height[i], left_max[i-1])
        
    right_max[n-1] = height[n-1]
    for i in range(n-2, -1, -1):
        right_max[i] = max(height[i], right_max[i+1])
        
    water = 0
    for i in range(n):
        water += min(left_max[i], right_max[i]) - height[i]
        
    return water

test_cases = [
    (([0,1,0,2,1,0,1,3,2,1,2,1],), 6),
    (([4,2,0,3,2,5],), 9),
    (([1,2,3,4,5],), 0),
    (([5,4,3,2,1],), 0)
]

run_tests(test_cases, trap_dp)
```

```python
# Optimal: Two Pointers 
# Time: O(n), Space: O(1)

def trap(height: list[int]) -> int:
    left, right = 0, len(height) - 1
    left_max = right_max = 0
    water = 0

    while left < right:
        if height[left] <= height[right]:
            # Right side is at least as tall — left_max determines water level
            left_max = max(left_max, height[left])
            water += left_max - height[left]
            left += 1
        else:
            # Left side is taller — right_max determines water level
            right_max = max(right_max, height[right])
            water += right_max - height[right]
            right -= 1

    return water

test_cases = [
    (([0,1,0,2,1,0,1,3,2,1,2,1],), 6),
    (([4,2,0,3,2,5],), 9),
    (([1,2,3,4,5],), 0),
    (([5,4,3,2,1],), 0)
]

run_tests(test_cases, trap)
```

### 📊 Complexity Analysis
- **Time Complexity:** O(n). We use two pointers that start at opposite ends and move inward until they meet. We process each element exactly once.
- **Space Complexity:** O(1). We strictly use a few integer tracking variables (`left`, `right`, `left_max`, `right_max`, `water`), completely eliminating the need for the O(n) DP prefix/suffix tracking arrays.

### 🎤 Interview Q&A
- **Q: Why does the two-pointer approach work?**
  A: When `height[left] <= height[right]`, the right boundary is guaranteed ≥ `left_max`. So water at `left` is exactly `left_max - height[left]`. We don't need to know the actual right max — just that it's ≥ left_max.
- **Q: What's the O(n) space solution?**
  A: Precompute `prefix_max` (left max at each index) and `suffix_max` (right max), then iterate once. Two pointers does the same in O(1) space.
- **Q: What happens if the array is continuously increasing or decreasing?**
  A: No water is trapped. The pointer on the shorter side will just continually move inward, updating its max but never finding a valley. For example, in `[1,2,3,4]`, left moves rightward, `left_max` updates, but difference is always 0.
- **Q: Could you solve this with a Monotonic Stack instead?**
  A: Yes. A decreasing monotonic stack can catch valleys and compute horizontally-bounded water pools when a taller wall is encountered. However, it takes O(n) space for the stack, making the Two Pointers approach strictly better for space.
- **Practice Problem Extension:** **LC #84 - Largest Rectangle in Histogram**. It flips the problem from dealing with bounding peaks to dealing with bounding floors, and highly leverages the Monotonic Stack concept.

### 💼 The Citi Angle
**Data Engineering: Windowed Capacity Headroom**
In data engineering for global transaction routing, message queues often suffer from bottleneck constraints generated by upstream ingestion bursts and downstream processing caps. Evaluating "water trapped" is physically equivalent to calculating the "Windowed Capacity Headroom" — how much temporary buffer memory is actually required to safely absorb spikes between variable structural limits. By knowing that the left-side ingestion firewall (`left_max`) is smaller than the right-side processing limit, we definitively know the queue will only fill up to the `left_max`. A two-pointer approach traverses millions of queue metrics in real-time O(n) without allocating heavy O(n) metric arrays, saving enormous memory overhead on the APM nodes.
