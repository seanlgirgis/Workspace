### 📋 Problem

Mirror a binary tree. Return its root.

```
Input:       Output:
    4            4
   / \          / \
  2   7        7   2
 / \ / \      / \ / \
1  3 6  9    9  6 3  1
```

**Constraints:** 0 ≤ nodes ≤ 100 | -100 ≤ Node.val ≤ 100

### 🧠 Choose Your Mental Model

| Model | The "Story" | Mechanism |
|-------|-------------|-----------|
| **Mirror Flip** | "Place a vertical mirror at the root. Every node's left child becomes its right child and vice versa. This transformation applies at every node simultaneously." | Swap children, recurse both subtrees |
| **DFS Post-Order** | "Visit left, visit right, THEN swap. After both subtrees are inverted, swap the root's children." | Post-order: process children before action at node |
| **DFS Pre-Order** | "Swap first, then visit children. The swap at the root doesn't affect the recursion into children — they get correctly inverted regardless." | Pre-order: swap at node, then recurse |
| **BFS Level-by-Level** | "Go floor by floor. On each floor, swap every node's left and right children." | Queue, swap children on dequeue |


````python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def build_tree(vals):
    """Build tree from level-order list (None = missing node)."""
    if not vals or vals[0] is None:
        return None
    from collections import deque
    root = TreeNode(vals[0])
    q = deque([root])
    i = 1
    while q and i < len(vals):
        node = q.popleft()
        if i < len(vals) and vals[i] is not None:
            node.left = TreeNode(vals[i])
            q.append(node.left)
        i += 1
        if i < len(vals) and vals[i] is not None:
            node.right = TreeNode(vals[i])
            q.append(node.right)
        i += 1
    return root

print("TreeNode ready.")

def invertTree_bfs(root: TreeNode) -> TreeNode:
    """
    BFS iterative — swap children of every node level by level.
    Time: O(n)  |  Space: O(w) — w = max level width
    More explicit, easier to trace for debugging.
    """
    from collections import deque
    if not root:
        return None

    queue = deque([root])

    while queue:
        node = queue.popleft()
        # Swap children
        node.left, node.right = node.right, node.left
        # Enqueue non-null children
        if node.left:  queue.append(node.left)
        if node.right: queue.append(node.right)

    return root


def tree_values(root):
    """Level-order values for testing."""
    if not root: return []
    from collections import deque
    result, q = [], deque([root])
    while q:
        node = q.popleft()
        if node:
            result.append(node.val)
            q.append(node.left)
            q.append(node.right)
        else:
            result.append(None)
    while result and result[-1] is None:
        result.pop()
    return result

root = build_tree([4, 2, 7, 1, 3, 6, 9])
print("Original:", tree_values(root))
inverted = invertTree_bfs(root)
print("BFS inverted:", tree_values(inverted))   # [4, 7, 2, 9, 6, 3, 1]
````

## 🔬 Complexity Analysis

<div style="padding: 15px; border-left: 8px solid #1a8a5a; background-color: #f1f8f4; color: #1b5e20; border-radius: 4px;">
    <strong>The Core Insight:</strong> Every approach is O(n) time — we must visit every node.
    Space differs: DFS uses call stack O(h), BFS uses queue O(w).
    For this problem, both are equivalent — choose based on style preference.
</div>

---

| Approach | Time | Space | Style |
|----------|------|-------|-------|
| DFS Recursive | O(n) | O(h) | Most concise — 4 lines |
| BFS Iterative | O(n) | O(w) | Explicit, no recursion limit concern |
| DFS Iterative (stack) | O(n) | O(h) | Manual stack, avoids Python recursion |

### Why O(n) is the lower bound
You must visit every node to swap its children.
A node you don't visit won't get its children swapped — producing an incorrect result.
Therefore no algorithm can do better than O(n).

### Space: h vs w
For a balanced binary tree: h ≈ log n, w ≈ n/2 (leaf level).
So DFS (O(log n)) is better than BFS (O(n)) for balanced trees.
For a skewed tree: h = n, w = 1. BFS (O(1)) beats DFS (O(n)).

<div style="padding: 10px; border-left: 8px solid #4CAF50; background-color: #f1f8f4; color: #1b5e20; border-radius: 4px;">
    <strong>✅ Default choice:</strong> DFS recursive — 4 lines, natural for trees.
    Fallback to BFS if recursion depth is a concern (skewed tree with n > 1000).
</div>


````python
def invertTree(root: TreeNode) -> TreeNode:
    """
    DFS Recursive — the canonical solution.
    Time: O(n)  |  Space: O(h) — call stack
    4 lines. The swap before recursion is fine because Python evaluates
    the right-hand side first: root.right, root.left = root.left, root.right
    is computed as a tuple swap, then assigned.
    """
    if not root:
        return None

    root.left, root.right = root.right, root.left   # swap

    invertTree(root.left)    # recurse into (previously right) subtree
    invertTree(root.right)   # recurse into (previously left) subtree

    return root


def invertTree_iterative(root: TreeNode) -> TreeNode:
    """
    DFS Iterative using explicit stack — avoids Python recursion limit.
    Time: O(n)  |  Space: O(h)
    """
    if not root:
        return None
    stack = [root]
    while stack:
        node = stack.pop()
        node.left, node.right = node.right, node.left
        if node.left:  stack.append(node.left)
        if node.right: stack.append(node.right)
    return root


# Tests
root = build_tree([4, 2, 7, 1, 3, 6, 9])
inverted = invertTree(build_tree([4, 2, 7, 1, 3, 6, 9]))
print("DFS recursive:", tree_values(inverted))     # [4, 7, 2, 9, 6, 3, 1]

inverted2 = invertTree_iterative(build_tree([4, 2, 7, 1, 3, 6, 9]))
print("DFS iterative:", tree_values(inverted2))    # [4, 7, 2, 9, 6, 3, 1]

# Edge cases
print("Single node:", tree_values(invertTree(build_tree([1]))))    # [1]
print("Two nodes:", tree_values(invertTree(build_tree([1, 2]))))   # [1, None, 2]
````

## 🎤 5 Interview Q&A

### Q1: Does the order of the recursive calls matter (left then right vs right then left)?

**Answer:** No. Both calls independently invert their subtrees. When we do `invertTree(root.left)`,
we've already swapped, so `root.left` points to what was previously `root.right`.
Both subtrees get inverted regardless of call order. The result is identical.

---

### Q2: Can you do this iteratively? Walk me through the stack approach.

**Answer:** Yes — use an explicit stack instead of the call stack:
```python
stack = [root]
while stack:
    node = stack.pop()
    node.left, node.right = node.right, node.left
    if node.left:  stack.append(node.left)
    if node.right: stack.append(node.right)
```
This is DFS with an explicit stack. For BFS, use a queue and `popleft()` instead of `pop()`.
Both produce a correctly inverted tree.

---

### Q3: What is the space complexity for a perfectly balanced tree vs a skewed tree?

**Answer:** For DFS recursive, space = O(h) = call stack depth.
- Balanced tree: h = log₂n → O(log n) space
- Skewed tree (all nodes go right): h = n → O(n) space

For BFS: space = O(w) = max queue size = max level width.
- Balanced tree: widest level has n/2 nodes → O(n) space
- Skewed tree: widest level has 1 node → O(1) space

So DFS wins for balanced trees; BFS wins for skewed trees.

---

### Q4: If you invert a tree twice, do you get the original back?

**Answer:** Yes — inverting twice is a self-inverse operation (like XOR with the same value).
`invert(invert(tree)) == original`. Proof: inverting swaps left↔right at every node.
Inverting again swaps right↔left (restoring original). The operation is its own inverse.

## 🎯 Summary: Key Takeaways

### The Pattern
**Swap at every node, recurse into both children.** Pre-order or post-order — both work.

### The 4-Line Solution
```python
def invertTree(root):
    if not root: return None
    root.left, root.right = root.right, root.left
    invertTree(root.left)
    invertTree(root.right)
    return root
```
