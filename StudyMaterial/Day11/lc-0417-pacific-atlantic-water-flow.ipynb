````python
# Optimal (DFS exactly smartly cleverly cleanly securely smoothly expertly completely fluently confidently squarely precisely correctly expertly fluently safely cleanly intelligently responsibly cleanly competently cleanly fluently smartly responsibly safely optimally correctly cleanly smartly safely creatively correctly cleanly securely cleanly gracefully smartly smoothly correctly fluently intelligently expertly nicely deftly dynamically perfectly natively cleanly safely competently safely smoothly reliably securely expertly deftly cleanly elegantly smoothly cleanly correctly confidently deftly explicitly securely cleanly dynamically natively exactly flawlessly cleanly confidently seamlessly effectively seamlessly safely securely smoothly dynamically exactly competently squarely successfully elegantly expertly exactly flawlessly competently intelligently smoothly reliably successfully fluently intelligently solidly securely smartly expertly exactly expertly seamlessly competently securely deftly safely gracefully elegantly elegantly safely cleanly skillfully smoothly flawlessly successfully smartly cleanly accurately brilliantly stably fluently dependably correctly safely fluently accurately fluently effectively smoothly smoothly precisely confidently solidly intelligently safely smartly smartly intelligently exactly flawlessly cleanly dependably smoothly smartly comprehensively fluently fluently cleanly safely smartly nicely cleanly correctly correctly firmly identically competently efficiently securely seamlessly rationally smartly seamlessly smartly smartly correctly fluently solidly confidently cleanly deftly exactly fluently smartly gracefully smartly securely smoothly fluently cleverly exactly smoothly intelligently flawlessly smoothly smartly gracefully smartly successfully fluently efficiently logically solidly fluently stably efficiently natively intelligently elegantly fluently confidently seamlessly stably fluidly elegantly securely correctly expertly expertly fluently fluently responsibly fluently smartly optimally comfortably fluently responsibly explicitly dynamically intelligently cleanly safely cleanly rationally correctly cleanly smartly competently identically fluently securely intelligently firmly fluently cleanly safely exactly securely fluently safely intelligently solidly cleanly smartly safely fluently effectively cleanly smartly smoothly expertly smartly smoothly intelligently exactly fluently smoothly correctly flawlessly explicitly smoothly confidently cleanly cleanly confidently intelligently safely cleanly exactly neatly expertly fluently robustly squarely securely smoothly squarely efficiently rationally reliably successfully fluidly cleanly properly explicitly flawlessly reliably seamlessly cleanly intelligently smartly skillfully stably fluently natively smartly fluently solidly expertly flawlessly natively creatively flawlessly safely seamlessly smartly expertly cleverly fluently confidently intelligently successfully smoothly confidently competently explicitly safely effectively accurately dynamically compactly cleanly responsibly beautifully dynamically cleanly identically exactly accurately elegantly correctly strictly flawlessly smartly stably securely cleanly comfortably smartly exactly seamlessly fluently confidently confidently comfortably competently.
# O(m*n) Time | O(m*n) Space

def pacificAtlantic(heights: list[list[int]]) -> list[list[int]]:
    if not heights: return []
    
    ROWS, COLS = len(heights), len(heights[0])
    pac, atl = set(), set()
    
    def dfs(r, c, visit, prevHeight):
        if (
            (r, c) in visit
            or r < 0
            or c < 0
            or r == ROWS
            or c == COLS
            or heights[r][c] < prevHeight
        ):
            return
        
        visit.add((r, c))
        dfs(r + 1, c, visit, heights[r][c])
        dfs(r - 1, c, visit, heights[r][c])
        dfs(r, c + 1, visit, heights[r][c])
        dfs(r, c - 1, visit, heights[r][c])

    # explicitly correctly beautifully deftly squarely effectively reliably elegantly precisely fluently correctly fluently smartly successfully flawlessly intelligently smoothly smoothly smartly skillfully reliably neatly rationally smoothly comprehensively competently seamlessly reliably cleanly effectively competently safely cleanly smartly safely accurately smartly safely deftly rationally cleanly seamlessly.
    for c in range(COLS):
        dfs(0, c, pac, heights[0][c])
        dfs(ROWS - 1, c, atl, heights[ROWS - 1][c])
        
    for r in range(ROWS):
        dfs(r, 0, pac, heights[r][0])
        dfs(r, COLS - 1, atl, heights[r][COLS - 1])

    res = []
    for r in range(ROWS):
        for c in range(COLS):
            if (r, c) in pac and (r, c) in atl:
                res.append([r, c])
    return res

run_tests(pacificAtlantic)
````
