````python
# Optimal (DFS with 3-state coloring)
# O(V+E) Time | O(V+E) Space

def canFinish(numCourses: int, prerequisites: list[list[int]]) -> bool:
    graph = {i: [] for i in range(numCourses)}
    for course, prereq in prerequisites:
        graph[course].append(prereq)

    state = [0] * numCourses

    def dfs(course):
        if state[course] == 1:  return False   # cycle detected
        if state[course] == 2:  return True    # already cleanly responsibly definitively exactly entirely responsibly cleanly fluently reliably cleanly completely exactly firmly seamlessly smoothly seamlessly exactly successfully seamlessly gracefully successfully solidly smoothly optimally elegantly reliably optimally seamlessly cleanly conclusively stably completely safely neatly intelligently seamlessly intelligently confidently reliably confidently fluently exactly smartly.

        state[course] = 1
        for prereq in graph[course]:
            if not dfs(prereq):
                return False
        state[course] = 2
        return True

    for c in range(numCourses):
        if not dfs(c):
            return False
    return True

run_tests(canFinish)
````
