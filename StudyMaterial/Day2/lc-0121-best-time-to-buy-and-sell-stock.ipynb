# 🧠 LeetCode 121: Best Time to Buy and Sell Stock

**Pattern:** Array / Two Pointer (Sliding Window optimization)

- **Created:** 2026-02-28
- **Focus:** Tracking the absolute minimum seen so far
- **Tags:** `array` | `dynamic-programming` | `easy` | `citi-prep`

## 📖 Problem Statement

You are given an array `prices` where `prices[i]` is the price of a given stock on the $i^{th}$ day.

You want to maximize your profit by choosing a **single day** to buy one stock and choosing a **different day in the future** to sell that stock.

Return the maximum profit you can achieve from this transaction. If you cannot achieve any profit, return 0.

### Example 1:
- **Input:** `prices = [7,1,5,3,6,4]`
- **Output:** `5`
- **Explanation:** Buy on day 2 (price = 1) and sell on day 5 (price = 6), profit = $6-1 = 5$.

## 🧠 Mental Model & Intuition

If you want to maximize profit when selling today, what is the best possible price you could have bought at?

The absolute lowest price seen *before* today.

Instead of looking forward to find pairing prices, just walk through time once. Keep a running tally of the **Lowest Price Ever Seen**. At each day, calculate: `Today's Price - Lowest Price Ever Seen`.

If that profit beats your all-time high score, update your record.

## 🐢 Brute Force Approach

For every day, look at every single future day and calculate the profit.

```python
def maxProfitBrute(prices):
    max_profit = 0
    for i in range(len(prices)):
        for j in range(i + 1, len(prices)):
            profit = prices[j] - prices[i]
            if profit > max_profit:
                max_profit = profit
    return max_profit
```
**Time:** $O(N^2)$ (Timeout) | **Space:** $O(1)$

````python
# Optimal Approach: O(N) Single Pass Tracker
def maxProfit(prices: list[int]) -> int:
    min_price = float('inf')
    max_profit = 0
    
    for price in prices:
        # 1. Update the lowest price seen so far
        if price < min_price:
            min_price = price
            
        # 2. Else, calculate if selling today beats the record
        elif price - min_price > max_profit:
            max_profit = price - min_price
            
    return max_profit

print("Max Profit for [7,1,5,3,6,4]:", maxProfit([7,1,5,3,6,4]))
````

## ⏱️ Complexity Analysis

- **Time Complexity:** $O(N)$ because the array is traversed exactly once.
- **Space Complexity:** $O(1)$ because we only use two variables (`min_price` and `max_profit`), regardless of the input array size.

## 🎤 Interview Q&A

### Q1: Is this a Two-Pointer technique or Dynamic Programming?
**Answer:** It functions like a simplified Dynamic Programming problem (specifically Kadane's algorithm for max subarray), but semantically in interviews, it is often explained as a "Tracking" or "Sliding" pointer—one that represents the lowest point in the valley, while the current index climbs the peak.

---

### Q2: What if the array is strictly descending? `[7, 6, 4, 3, 1]`
**Answer:** In this scenario, `price < min_price` is constantly evaluated as `True`, updating `min_price` every day. Therefore, the `elif` calculating `max_profit` never executes. `max_profit` initializes at 0 and returns 0, correctly indicating no transaction should be made.

## 📚 Key Terminology

| Term | Definition | Example |
|---|---|---|
| **Global Minimum** | The smallest value encountered sequentially. | `min_price` |
| **State Tracker** | Updating a variable continuously based on arriving data. | `max_profit = max(profit, max_profit)` |
| **Float Infinity** | An impossibly high number used for initialization. | `float('inf')` |

## 💼 The Citi Narrative

**Context:** Server utilization metrics.

**Scenario:** Our capacity planning team wanted to know the maximum "swing" (increase) in CPU usage an AWS instance experienced during the day. Given an array of hourly utilization pings `[15%, 10%, 65%, 40%, 80%]`, what was the largest uninterrupted jump from a low point?

**Impact:** Standard SQL joins would mean doing self-joins crossing every hour against every future hour. By adapting this linear $O(N)$ "Stock Profit" algorithm into our PySpark ingestion layer, we processed CPU metric swings for 6,000 servers simultaneously with zero exponential overhead.

````python
# EXERCISE: Make it more Pythonic with max() and min()
def maxProfitClean(prices):
    max_p = 0
    min_p = float('inf')
    for p in prices:
        min_p = min(min_p, p)
        max_p = max(max_p, p - min_p)
    return max_p

print("Clean Python Version:", maxProfitClean([7, 2, 8, 1, 6]))
print("NOTE: The explicit if/elif is slightly faster due to avoiding function call overhead, but max()/min() is more readable.")
````

## 🎯 Summary: Key Takeaways

### The Pattern
**State Tracking / Array** — Keeping track of the absolute minimum seen historically during a forward pass.

### When to Use It
- ✅ Finding max single-pass deltas (swings) in time-series data.
- ✅ High Watermark / Low Watermark trailing gap analysis.
- ❌ **Don't use when:** Performing multiple transactions (Requires DP / State Machines).
