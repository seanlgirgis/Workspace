# 🧠 LeetCode 3: Longest Substring Without Repeating Characters

**Pattern:** Sliding Window

- **Created:** 2026-02-28
- **Focus:** Expanding and contracting a window to eliminate duplicates
- **Tags:** `string` | `sliding-window` | `hash-set` | `medium` | `citi-prep`

## 📖 Problem Statement

Given a string `s`, find the length of the **longest substring** without repeating characters.

A **substring** is a contiguous non-empty sequence of characters within a string.

### Example 1:
- **Input:** `s = "abcabcbb"`
- **Output:** `3`
- **Explanation:** The answer is "abc", with the length of 3.

## 🧠 Mental Model & Intuition

Imagine looking through a rigid picture frame at a line of letters. 

- Stretch the right side of the frame out rightwards to capture more letters.
- We use a Hash Set as a "cheat sheet" to track what's currently inside the frame.
- Oh no! The next letter `c` is already in our set! We have a duplicate.
- So, we push the *left* side of the frame rightwards until we kick the original `c` out of the frame.
- Then we continue stretching the right side.

This "stretchy accordion" frame is called a **Sliding Window**.

## 🐢 Brute Force Approach

Evaluate every possible substring. Is it unique? Track its length.

```python
def lengthOfLongestSubstringBrute(s):
    max_len = 0
    for i in range(len(s)):
        for j in range(i + 1, len(s) + 1):
            sub = s[i:j]
            if len(set(sub)) == len(sub):  # Check if unique
                max_len = max(max_len, len(sub))
    return max_len
```
**Time:** $O(N^3)$ (Two loops + the $O(N)$ string slicing/set creation) | **Space:** $O(N)$

````python
# Optimal Sliding Window Approach
def lengthOfLongestSubstring(s: str) -> int:
    char_set = set() # To store characters currently in our window
    l = 0            # Left pointer
    max_len = 0      # Record tracker
    
    for r in range(len(s)): # r is the Right pointer
        # If the character at r is already in the set, we must shrink the window from the left.
        while s[r] in char_set:
            # Remove the left-most character from the set
            char_set.remove(s[l])
            # Move the left pointer over by 1
            l += 1
            
        # Now the window is safe. Add the new right-most character.
        char_set.add(s[r])
        
        # Calculate current window size
        window_size = r - l + 1
        max_len = max(max_len, window_size)
        
    return max_len

print("Longest for 'abcabcbb':", lengthOfLongestSubstring("abcabcbb"))
print("Longest for 'pwwkew':", lengthOfLongestSubstring("pwwkew"))
````

## ⏱️ Complexity Analysis

- **Time Complexity:** $O(N)$. Even with the `while` loop inside the `for` loop, each character is added to the set exactly once, and removed exactly once. Thus, $2N$ overall operations, which simplifies to $O(N)$.
- **Space Complexity:** $O(min(N, M))$ where $M$ is the size of the alphabet (e.g., 26 for English letters, or 128/256 for ASCII). The set will never grow larger than the character set.

## 🎤 Interview Q&A

### Q1: Is there a way to avoid the inner `while` loop that increments `l` step-by-step?
**Answer:** Yes! Instead of a Hash *Set*, use a Hash *Map* tracking the exact indices of each character observed. If you hit a duplicate, you can instantly warp the `l` pointer to `max(l, last_seen_index + 1)`. This is slightly more complex to write, but theoretically faster as it skips intermediate character removals.

---

### Q2: Why is the condition in the while loop `s[r] in char_set` and not `len(char_set) == window_size`?
**Answer:** It's structurally identical logic, but `in` checks against the HashSet in $O(1)$ time and is semantically clear: "while the bad character is still in my bucket, keep throwing out left-side characters."

## 📚 Key Terminology

| Term | Definition | Example |
|---|---|---|
| **Sliding Window** | Establishing specific bounds `[left, right]` and adjusting them to satisfy constraints. | `r - l + 1` |
| **Hash Set** | Data structure to check uniqueness in O(1) time. | `set()` |
| **Contiguous Sequence** | Elements must be adjacent in memory next to each other. | A substring, as opposed to a subsequence. |

## 💼 The Citi Narrative

**Context:** Network Packet Deduplication.

**Scenario:** In high-speed trading, market feeds emit contiguous streams of network packets. If our network card jittered, we might receive duplicate message IDs within a short window. We needed to identify the longest "clean" streak of unique messages to evaluate network stability.

**Impact:** Doing an $O(N^2)$ cross query on our Hadoop cluster to identify distinct unbroken sequences took hours. Implementing an algorithm based on this exact $O(N)$ sliding window allowed us to process the stream functionally in real-time as the packets arrived.

````python
# EXERCISE: Implement the O(N) Hash Map "Warp" optimization
def lengthWarpAlgorithm(s: str) -> int:
    char_map = {}
    l = 0
    max_len = 0
    
    for r in range(len(s)):
        if s[r] in char_map:
            # Warp `l` forward past the previous occurrence.
            # Use `max` to prevent warp loops from pulling `l` backwards.
            l = max(l, char_map[s[r]] + 1)
            
        char_map[s[r]] = r
        max_len = max(max_len, r - l + 1)
        
    return max_len

print("Warp algorithm result:", lengthWarpAlgorithm("abba"))
````

## 🎯 Summary: Key Takeaways

### The Pattern
**Sliding Window** — Expanding a right pointer and contracting a left pointer to bypass invalid states.

### When to Use It
- ✅ Finding longest contiguous sequence meeting conditions.
- ✅ Deduplication applied to order-dependent streaming packets.
- ❌ **Don't use when:** Problem requires non-contiguous elements (Use subsequences/DP).
